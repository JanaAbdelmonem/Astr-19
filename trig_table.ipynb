# Sin & Cos Description
Here we're defining functions for sin(x) and cos(x) then putting their values in a table for x between 0 and 2.

```python
import numpy as np

def sin_function(x):
    """
    Returns sin of x.
    """
    return np.sin(x)
```

## sin(x) Function

This function will take a value(s) `x` and return the
sin of `x` using numpy.


```python
def cos_function(x):
    """
    Returns cos of x.
    """
    return np.cos(x)
```

## cos(x) Function

This function will take a value(s) `x` and return the
cos of `x` using numpy.


```python
# create x values
x = np.linspace(0.0, 2.0, 1000)

# turn values to a table
sin_values = sin_function(x)
cos_values = cos_function(x)
```

## The table of sin(x) and cos(x)

The next cell will create 1000 evenly spaced values of `x` between
0 and 2, then will calculate and find the values of sin(x) and cos(x) using the previously defined
functions.


```python
print("    x        sin(x)     cos(x)")
print("--------------------------------")

for i in range(10):
    print(f"{x[i]:8.5f}  {sin_values[i]:8.5f}  {cos_values[i]:8.5f}")
```
