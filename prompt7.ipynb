```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 1, 100)

def exp_function(x):
    return np.exp(x)
```

```python
y = exp_function(x)

plt.figure()
plt.plot(x, y)

plt.xlabel("Time [millieseconds]")
plt.ylabel("Awesomeness")

plt.savefig("plot.pdf")
plt.show()
```
